# changing data to be dataframe

```python
df=pd.DataFrame(np.c_[cancer["data"], cancer["target"]] , columns=np.append(cancer["feature_names"] , "target"))
df
```

```python
df.info()
```

```python
df["target"]=df["target"].astype("int")
```

```python
data=df.copy()
```

```python
inputs=data.drop("target", axis=1)
target=data["target"]
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train , y_test=train_test_split(inputs , target , test_size=0.2 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler 

scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# ML algorithems

# SVC(support vector claassidy)

```python
from sklearn.model_selection import cross_validate
from sklearn.svm import SVC

scoring = ['precision_macro', 'recall_macro']

svc=SVC()
scores = cross_validate(svc,x_train_scaled , y_train, scoring=scoring)
svc.fit(x_train_scaled ,y_train)       
svc_pred=svc.predict(x_test_scaled)
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(svc , x_test_scaled , y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,svc_pred)
```

```python
error=1-accuracy_score(y_test,svc_pred)
error
```

# improving SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid={"C":[0.01 , 0.1 ,1] , "kernel":["libnear","rbf" ,"sigmoid"] , "degree":[2,3,4,5],"gamma":["scale","auto"]}
svc_grid=GridSearchCV(svc , param_grid)
svc_grid.fit(x_train_scaled , y_train)
```

```python
svc_grid.best_params_
```

```python
svc_grid_pred=svc_grid.predict(x_test_scaled)
```

```python
accuracy_score(y_test,svc_grid_pred)
```

```python
error=1-accuracy_score(y_test,svc_grid_pred)
error
```

```python
sns.kdeplot(y_test - svc_pred )
plt.title("residual" , size=15)
```
