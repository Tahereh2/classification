```python
#df["Gender"]=df["Gender"].map({"Male":0 ,"Female":1})
df["Gender"]=df["Gender"].replace("Male",0)
df["Gender"]=df["Gender"].replace("Female",1)
```

```python
df["EverBenched"]=df["EverBenched"].replace("No",0)
df["EverBenched"]=df["EverBenched"].replace("Yes",1)
```

```python
df_city_1hot=pd.get_dummies(df["City"] , drop_first=True)
```

```python
df_city_1hot.head()
```

```python
df=df.drop("City" , axis=1)
```

```python
df=pd.concat([df , df_city_1hot], axis=1)
```

```python
df.head()
```

```python
from category_encoders.ordinal import OrdinalEncoder

ord_cols_mapping=[{"col":"Education"  ,"mapping":{"Bachelors":0 ,"Masters":1 ,"PHD":2}}]
```

```python
encoder=OrdinalEncoder(mapping=ord_cols_mapping)
df=encoder.fit_transform(df)
```

```python
df.head()
```

```python
inputs=df.drop("LeaveOrNot", axis=1)
target=df["LeaveOrNot"]
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train , y_test=train_test_split(inputs , target , test_size=0.25 , random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier(random_state=42 ,max_depth=5 ,max_leaf_nodes=8) 
dtc.fit(x_train , y_train)
```

```python
dtc_pred=dtc.predict(x_test)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8) , dpi=200)
plot_tree(dtc);
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , dtc_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(dtc ,x_test , y_test)
```

```python
from sklearn.model_selection import GridSearchCV

dt=DecisionTreeClassifier()
param_grid={"max_depth":[2,3,4,5, 6, 7 ,8] , "max_leaf_nodes":[7,8,9,10],"criterion":["gini","entropy","log_loss"] ,"min_samples_split":[1,2,3,3]}
grid=GridSearchCV(dt , param_grid)
grid.fit(x_train , y_train)
```

```python
grid.best_params_
```

```python
dt_grid_pred=grid.predict(x_test)
```

```python
dt=DecisionTreeClassifier(criterion= 'gini',max_depth= 6,max_leaf_nodes= 10,min_samples_split= 2)
dt.fit(x_train , y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8) , dpi=200)
plot_tree(dt);
```

```python
print(classification_report(y_test , dt_grid_pred))
```
