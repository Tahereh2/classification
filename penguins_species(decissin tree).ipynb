```python
inputs=pd.get_dummies(df.drop("species" , axis=1) , drop_first=True)
target=df["species"]
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train , y_test=train_test_split(inputs, target , test_size=0.2 , random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
dtc.fit(x_train , y_train)
dtc_pred=dtc.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , dtc_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(dtc , x_test , y_test)
```

```python
dtc.feature_importances_
```

```python
pd.DataFrame(index=inputs.columns , data=dtc.feature_importances_ , columns=["feature_importances"]).sort_values("feature_importances")
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8) , dpi=200)
plot_tree(dtc);
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

log=LogisticRegression(multi_class="ovr")
log.fit(x_train , y_train)                                        

log_pred=log.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , log_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(log , x_test , y_test)
```
